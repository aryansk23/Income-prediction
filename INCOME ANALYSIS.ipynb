# Treating the error values

```python
df['workclass'].replace(' ?', np.nan, inplace = True)
```

```python
df['workclass'].unique()
```

```python
df['workclass'].isnull().sum()
```

```python
df['workclass'].fillna(df['workclass'].mode()[0], inplace = True)
```

```python
df['workclass'].unique()
```

```python
df['workclass'].isnull().sum()
```

```python
df['occupation'].replace(' ?', np.nan, inplace = True)
```

```python
df['occupation'].fillna(df['occupation'].mode()[0], inplace = True)
```

```python
df['occupation'].unique()
```

```python
df['native-country'].replace(' ?', np.nan, inplace = True)
```

```python
df['native-country'].fillna(df['native-country'].mode()[0], inplace = True)
```

```python
df['native-country'].unique()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df['Income'].unique()
```

```python
dff = df.groupby('workclass')['Income'].value_counts()
```

```python
plt.figure(figsize = (25,35))
dff.plot(kind = 'pie', autopct = '%1.1f%%', startangle = 45, cmap = 'inferno')
plt.title('Work class by income')
plt.ylabel('pie chart')
plt.legend(title = 'Salary by job sector')
plt.show()
```

### here we can see the diffrent workclass with their approximate income values (' <=50K', ' >50K')

```python
education_and_income = df.groupby('education')['capital-gain'].sum()
```

```python
plt.figure(figsize = (15,5))
education_and_income.sort_values(ascending= False).plot(kind='bar', color ='blue')

plt.show()
```

### The graph shows us the education level with the capital gain they were receiving

```python
occupation_capital_gain = df.groupby('occupation')['capital-gain'].sum()
```

```python
plt.figure(figsize = (15,10))
occupation_capital_gain.plot(kind = 'pie',autopct = '%1.1f%%',cmap = 'viridis' )
plt.title('Occupation and Capital gain')
plt.ylabel('pie chart')
plt.legend(title = 'Capital Gain')
plt.show()
```

### Exec_managerial is receiving the highest capital gain and priv-house-service is receiving the lowest capital gain

```python
hours_per_week = df.groupby('race')['hours-per-week'].sum()
```

```python
plt.figure(figsize = (15,5))
hours_per_week.sort_values(ascending= False).plot(kind='line', color ='blue')

plt.show()
```

### White race has the highest working hours
